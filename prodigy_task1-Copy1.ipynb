```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('population.csv',  encoding='utf-8')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
df_final.head()
```

```python

years = [str(year) for year in range(1960, 2024)]
population_data = df[['Country Name'] + years].copy()

# Sum the population data across the years for each country
population_data['Total Population (1960-2023)'] = population_data[years].sum(axis=1)

# Sort by total population for better visualization
population_data_sorted = population_data.sort_values(by='Total Population (1960-2023)', ascending=False)

# Plotting the bar chart for the top 20 countries by total population from 2000 to 2023
top_20_countries = population_data_sorted.head(20)

plt.figure(figsize=(12, 8))
plt.barh(top_20_countries['Country Name'], top_20_countries['Total Population (1960-2023)'], color='skyblue')
plt.xlabel('Total Population (1960-2023)')
plt.ylabel('Country')
plt.title('Top 20 Countries by Total Population (1960-2023)')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest population at the top
plt.show()
```

```python
chosen_country = 'India'


country_data = df_final[df_final['Country Name'] == chosen_country]


years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')


plt.show()
```

```python
plt.hist(population, bins=20, color='yellow', edgecolor='black')
plt.show()
```
